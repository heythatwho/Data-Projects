```python
import pandas as pd
from sqlalchemy import create_engine
import os

# Database credentials
db_username = 'postgres'
db_password = '123123123'
db_host = 'localhost'
db_port = '5432'
db_name = 'myfirstpostgresqldb'

# Create the SQLAlchemy engine
engine = create_engine(f'postgresql+psycopg2://{db_username}:{db_password}@{db_host}:{db_port}/{db_name}')

# Directory containing the files
file_directory = '/Users/issacmiao/Downloads/archive (1)'

# Check if the directory exists
if not os.path.exists(file_directory):
    print(f"Directory {file_directory} does not exist.")
else:
    print(f"Processing files from directory: {file_directory}")

# Iterate through all files in the directory
for filename in os.listdir(file_directory):
    file_path = os.path.join(file_directory, filename)
    
    if filename.endswith('.csv') or filename.endswith('.xlsx'):
        print(f"Processing file: {file_path}")
        
        try:
            # Load the file into a DataFrame
            if filename.endswith('.csv'):
                df = pd.read_csv(file_path, encoding='ISO-8859-1')
                print(f"Successfully read {filename} as CSV")
            elif filename.endswith('.xlsx'):
                df = pd.read_excel(file_path)
                print(f"Successfully read {filename} as Excel")
            
            # Define the table name (remove file extension for table name)
            table_name = os.path.splitext(filename)[0]
            
            # Insert the DataFrame into the PostgreSQL table
            df.to_sql(table_name, engine, if_exists='replace', index=False)
            print(f"Added {table_name} to the database.")
        
        except Exception as e:
            print(f"Error processing {file_path}: {e}")

print("Finished processing files.")
```

```python
import pandas as pd
import psycopg2
import matplotlib.pyplot as plt

# Connect to PostgreSQL database
conn = psycopg2.connect(
    host="localhost",
    database="postgres",
    user="postgres",
    password="123123123"
)
cur = conn.cursor()

# Load CSV data into a pandas dataframe with encoding
csv_file = '/Users/issacmiao/Downloads/sales_data.csv'
df = pd.read_csv(csv_file, encoding='utf-8')  # Specify encoding here

# Clean the data
df.columns = df.columns.str.strip()  # Clean up whitespace
df.dropna(inplace=True)  # Remove missing values
df.drop_duplicates(inplace=True)  # Remove duplicates

# Convert sale_date to datetime
df['sale_date'] = pd.to_datetime(df['Order Date'])

# Check for outliers
df = df[(df['Quantity Ordered'] > 0) & (df['Sales'] > 0)]

# Create new calculated fields
df['total_profit'] = df['Sales'] - (df['Quantity Ordered'] * 0.5)  # Assume a cost price of 0.5

# Categorize products based on sales
sales_bins = [0, 100, 500, 1000, float('inf')]
sales_labels = ['Low', 'Medium', 'High', 'Very High']
df['sales_category'] = pd.cut(df['Sales'], bins=sales_bins, labels=sales_labels)

# Analyze the data
total_sales_by_product = df.groupby('Product')['Sales'].sum().reset_index()

# Visualize total sales by product
plt.figure(figsize=(12, 6))
plt.bar(total_sales_by_product['Product'], total_sales_by_product['Sales'])
plt.xlabel('Product')
plt.ylabel('Total Sales')
plt.title('Total Sales by Product')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Create cleaned and transformed table in PostgreSQL
create_cleaned_table_query = '''
CREATE TABLE IF NOT EXISTS sales_transformed_data (
    id SERIAL PRIMARY KEY,
    sale_date TIMESTAMP,
    product_name VARCHAR(255),
    quantity_sold INTEGER,
    price NUMERIC,
    total_amount NUMERIC,
    total_profit NUMERIC,
    sales_category VARCHAR(50)
);
'''
cur.execute(create_cleaned_table_query)
conn.commit()

# Insert cleaned and transformed data into PostgreSQL
for _, row in df.iterrows():
    insert_query = '''
    INSERT INTO sales_transformed_data (sale_date, product_name, quantity_sold, price, total_amount, total_profit, sales_category)
    VALUES (%s, %s, %s, %s, %s, %s, %s)
    '''
    # Encoding the data before insertion
    cur.execute(insert_query, (row['sale_date'], row['Product'], row['Quantity Ordered'], row['Price Each'], row['Sales'], row['total_profit'], row['sales_category']))

conn.commit()

# Clean up
cur.close()
conn.close()
```
